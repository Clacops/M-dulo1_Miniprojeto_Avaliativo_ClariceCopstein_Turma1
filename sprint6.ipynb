## 📌 1. Introdução e Objetivo do Projeto

O objetivo principal foi realizar o pipeline completo de engenharia, saneamento, exploração estatística e extração de inteligência de negócios (*Business Intelligence*) sobre uma base de dados histórica transacional de varejo, contendo originalmente **830.000 registros**.

## ⚙️ 2. Metodologia de Desenvolvimento (Pipeline em Sprints)

O projeto foi segmentado de forma incremental e estruturada para garantir a governança e a qualidade dos dados:

* **Sprint 1 & 2: Carga e Saneamento de Estruturas**
    * Leitura otimizada do ecossistema de arquivos e acoplamento das bases transacionais.
    * Tratamento cronológico da variável de tempo (`DATA`), convertendo strings brutas para o formato nativo de *datetime* do Python.
* **Sprint 3: Purificação e Auditoria de Regras de Negócio**
    * *Remoção de Duplicados:* Identificação e eliminação de registros idênticos para evitar a inflação artificial do volume de vendas.
    * *Tratamento de Não-Conformidades (#N/D):* Investigação profunda e eliminação de falhas de sistema nas colunas `PR_CAT` e `PR_NOME`. 
    * *Nota de Auditoria:* Durante esta etapa, foi realizado um cruzamento de integridade com o Microsoft Excel, detectando que das 3.650 ocorrências originais de erro, 422 também eram duplicadas, restando 3.228 erros únicos que foram purificados via Pandas.
* **Sprint 4: Estatística Descritiva Univariada**
    * Exploração matemática e cálculo de medidas de tendência central, dispersão e limites sobre a variável de dependentes do cliente (`CL_FHL`).
    * Geração e exportação física de tabelas de frequência (CSV) e plots de distribuição (PNG).
* **Sprint 5: Análise Multivariada e Padrões de Consumo (BI)**
    * Cruzamento de dados categóricos e temporais para mapeamento de comportamento de compra.
    * Geração de matrizes estatísticas multidimensionais e mapas de calor.
