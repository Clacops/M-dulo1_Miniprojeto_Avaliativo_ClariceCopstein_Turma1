```python
import pandas as pd
from rich import print
from rich.rule import Rule
from rich.panel import Panel

print(Rule("SPRINT 1: IMPORTAÇÃO E DIAGNÓSTICO DOS DADOS", style="bold cyan"))

# 01. Importando a tabela
print("\n[bold yellow][1] Importando a base de dados...[/bold yellow]")
df = pd.read_csv('Base Varejo.csv', sep=';') 

# 02. Definindo a apresentação das tabelas e casas decimais
print("[bold yellow][2] Configurando a apresentação do Pandas...[/bold yellow]")
pd.set_option('display.max_columns', None)       
pd.set_option('display.max_rows', 20)            
pd.set_option('display.float_format', '{:.2f}'.format) 

# 03. Quantidade de linhas e colunas - tipos de dados
print("\n[bold yellow][3] Diagnóstico Inicial:[/bold yellow]")

# Criando um texto formatado para o Painel
texto_dimensoes = f"-> Total de Registros (Linhas): [bold green]{df.shape[0]}[/bold green]\n-> Total de Atributos (Colunas): [bold green]{df.shape[1]}[/bold green]"
print(Panel(texto_dimensoes, title="Dimensões da Tabela", border_style="cyan"))

print("\n[bold underline]Tipos de Dados por Coluna:[/bold underline]")
print(df.dtypes)

# 04. Salvando o arquivo com o novo nome e o novo separador
print("\n[bold yellow][4] Exportando a nova base...[/bold yellow]")
novo_nome = 'base_varejo_sprint1.csv'
df.to_csv(novo_nome, sep=',', index=False)

print(Rule(f"[bold green][OK] Sprint 1 concluído! Arquivo guardado como '{novo_nome}'.[/bold green]", style="bold green"))
print("\n")
```
